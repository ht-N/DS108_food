```python
# Remove rows with any missing values (as specified in the instructions)
df_clean = df.dropna(axis=0)
print(f"Original dataset shape: {df.shape}")
print(f"Cleaned dataset shape: {df_clean.shape}")
print(f"Rows removed: {df.shape[0] - df_clean.shape[0]}")

# Use cleaned dataset for further analysis
df = df_clean.copy()
```

```python
# Convert numeric salary columns
# Handle 'TBD' values in salary columns
df['minSalary_numeric'] = pd.to_numeric(df['minSalary'], errors='coerce')
df['maxSalary_numeric'] = pd.to_numeric(df['maxSalary'], errors='coerce')

# Create average salary column
df['avgSalary'] = (df['minSalary_numeric'] + df['maxSalary_numeric']) / 2

print("Salary conversion completed.")
print(f"Rows with valid salary data: {df['avgSalary'].notna().sum()}")
```

```python
# Count job distribution by location
location_counts = df['Location'].value_counts()

# Filter locations with count > 0 and get top 15
top_locations = location_counts[location_counts > 0].head(15)

print("Top 15 locations by job count:")
print(top_locations)
```

```python
# Create horizontal bar chart for job distribution by location
plt.figure(figsize=(12, 8))
bars = plt.barh(top_locations.index, top_locations.values, color='skyblue', edgecolor='navy')

# Add value labels
for bar in bars:
    width = bar.get_width()
    plt.text(width + 1, bar.get_y() + bar.get_height()/2, 
             f'{int(width)}', ha='left', va='center', fontweight='bold')

plt.title('Distribution of Jobs by Location (Top 15)', fontsize=16, fontweight='bold')
plt.xlabel('Number of Jobs', fontsize=14)
plt.ylabel('Location', fontsize=14)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Function to categorize experience
def categorize_experience(exp):
    if pd.isna(exp):
        return 'N/A'
    if exp == 0:
        return 'No Experience'
    if exp < 1:
        return '< 1 year'
    if exp < 2:
        return '1-2 years'
    if exp < 3:
        return '2-3 years'
    if exp < 5:
        return '3-5 years'
    return '5+ years'

# Apply categorization
df['Experience_Category'] = df['Experience_year'].apply(categorize_experience)
experience_counts = df['Experience_Category'].value_counts()

print("Experience distribution:")
print(experience_counts)
```

```python
# Create pie chart for experience distribution
plt.figure(figsize=(10, 8))
colors = sns.color_palette("Set3", len(experience_counts))

plt.pie(experience_counts.values, 
        labels=experience_counts.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors,
        explode=[0.05 if i == experience_counts.values.argmax() else 0 for i in range(len(experience_counts))])

plt.title('Distribution of Jobs by Experience Requirements', fontsize=16, fontweight='bold')
plt.axis('equal')

# Add legend
plt.legend(experience_counts.index, 
          title="Experience Level",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

plt.tight_layout()
plt.show()
```

```python
# Define salary ranges for binning
salary_bins = [0, 10, 20, 30, 40, 50, 100, float('inf')]
salary_labels = ['0-10M', '10-20M', '20-30M', '30-40M', '40-50M', '50-100M', '100M+']

# Filter out NaN salaries for analysis
df_salary = df.dropna(subset=['avgSalary'])

# Categorize salaries
df_salary['Salary_Range'] = pd.cut(df_salary['avgSalary'], bins=salary_bins, labels=salary_labels)

# Count occurrences
salary_counts = df_salary['Salary_Range'].value_counts().sort_index()

print("Salary distribution:")
print(salary_counts)
```

```python
# Create bar plot for salary distribution
plt.figure(figsize=(12, 6))
bars = plt.bar(salary_counts.index, salary_counts.values, 
               color='lightcoral', edgecolor='darkred', alpha=0.8)

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.5, 
             f'{int(height)}', ha='center', va='bottom', fontweight='bold')

plt.title('Distribution of Jobs by Salary Range (VND Million)', fontsize=16, fontweight='bold')
plt.xlabel('Salary Range (VND)', fontsize=14)
plt.ylabel('Number of Jobs', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Count company size distribution
company_size_counts = df['Company Size'].value_counts()

print("Company size distribution:")
print(company_size_counts)
```

```python
# Create bar plot for company size distribution
plt.figure(figsize=(12, 6))
bars = plt.bar(company_size_counts.index, company_size_counts.values, 
               color='lightcoral', edgecolor='darkred', alpha=0.8)

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 1, 
             f'{int(height)}', ha='center', va='bottom', fontweight='bold')

plt.title('Distribution of Jobs by Company Size', fontsize=16, fontweight='bold')
plt.xlabel('Company Size (Number of Employees)', fontsize=14)
plt.ylabel('Number of Jobs', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
print("=== KEY INSIGHTS FROM EDA ===")
print("\n1. DATASET OVERVIEW:")
print(f"   - Total jobs analyzed: {len(df):,}")
print(f"   - Number of fields/industries: {df['Field'].nunique()}")
print(f"   - Number of locations: {df['Location'].nunique()}")
print(f"   - Jobs with salary information: {df['avgSalary'].notna().sum():,}")

print("\n2. LOCATION INSIGHTS:")
top_3_locations = df['Location'].value_counts().head(3)
for i, (location, count) in enumerate(top_3_locations.items(), 1):
    percentage = (count / len(df)) * 100
    print(f"   {i}. {location}: {count:,} jobs ({percentage:.1f}%)")

print("\n3. EXPERIENCE INSIGHTS:")
exp_summary = df['Experience_Category'].value_counts()
most_common_exp = exp_summary.index[0]
print(f"   - Most common experience requirement: {most_common_exp}")
print(f"   - Jobs requiring no experience: {exp_summary.get('No Experience', 0)}")

print("\n4. SALARY INSIGHTS:")
salary_data = df.dropna(subset=['avgSalary'])
if len(salary_data) > 0:
    print(f"   - Average salary: {salary_data['avgSalary'].mean():.1f} million VND")
    print(f"   - Median salary: {salary_data['avgSalary'].median():.1f} million VND")
    print(f"   - Salary range: {salary_data['avgSalary'].min():.1f} - {salary_data['avgSalary'].max():.1f} million VND")

print("\n5. INDUSTRY INSIGHTS:")
top_3_fields = df['Field'].value_counts().head(3)
for i, (field, count) in enumerate(top_3_fields.items(), 1):
    percentage = (count / len(df)) * 100
    print(f"   {i}. {field}: {count:,} jobs ({percentage:.1f}%)")

print("\n6. COMPANY SIZE INSIGHTS:")
company_size_summary = df['Company Size'].value_counts()
most_common_size = company_size_summary.index[0]
print(f"   - Most common company size: {most_common_size}")
print(f"   - Number of different company size categories: {len(company_size_summary)}")
```
